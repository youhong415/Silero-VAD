# librosa & silero 基本

```python
import librosa
import matplotlib.pyplot as plt
# 加载音频文件
audio_path = 'en_example.wav'  # 替换为您的音频文件路径
y, sr = librosa.load(audio_path)
```

```python
import torch
from pprint import pprint
import numpy as np
import time

SAMPLING_RATE = 16000
torch.set_num_threads(1)
# download example
#torch.hub.download_url_to_file('https://models.silero.ai/vad_models/en.wav', 'en_example.wav')
```

# ONNX 運算時間節省很多! (而且Probability看起來差不多!)

```python
# 开始计时
start_time = time.time()

USE_ONNX = True # change this to True if you want to test onnx model
model, utils = torch.hub.load(repo_or_dir='./snakers4_silero-vad_master',
                              source='local',
                              model='silero_vad',
                              force_reload=True,
                              onnx=USE_ONNX)

(get_speech_timestamps,
 save_audio,
 read_audio,
 VADIterator,
 collect_chunks) = utils

# 假设y是音频数据，wav是用于VAD的音频，model是VAD模型
# 首先获取语音时间戳和概率
wav = read_audio('en_example.wav', sampling_rate=SAMPLING_RATE)
speech_probs = []
window_size_samples = 512 # use 256 for 8000 Hz model
for i in range(0, len(wav), window_size_samples):
    chunk = wav[i: i+window_size_samples]
    if len(chunk) < window_size_samples:
        break
    speech_prob = model(chunk, SAMPLING_RATE).item()
    speech_probs.append(speech_prob)
model.reset_states() # reset model states after each audio

speech_onnx_true = speech_probs

# 结束计时
end_time = time.time()
execution_time = end_time - start_time
print(f"执行时间：{execution_time} 秒")
```

```python
# 开始计时
start_time = time.time()

USE_ONNX = False # change this to True if you want to test onnx model
model, utils = torch.hub.load(repo_or_dir='./snakers4_silero-vad_master',
                              source='local',
                              model='silero_vad',
                              force_reload=True,
                              onnx=USE_ONNX)

(get_speech_timestamps,
 save_audio,
 read_audio,
 VADIterator,
 collect_chunks) = utils

# 假设y是音频数据，wav是用于VAD的音频，model是VAD模型
# 首先获取语音时间戳和概率
wav = read_audio('en_example.wav', sampling_rate=SAMPLING_RATE)
speech_probs = []
window_size_samples = 512 # use 256 for 8000 Hz model
for i in range(0, len(wav), window_size_samples):
    chunk = wav[i: i+window_size_samples]
    if len(chunk) < window_size_samples:
        break
    speech_prob = model(chunk, SAMPLING_RATE).item()
    speech_probs.append(speech_prob)
model.reset_states() # reset model states after each audio

speech_onnx_false = speech_probs

# 结束计时
end_time = time.time()
execution_time = end_time - start_time
print(f"执行时间：{execution_time} 秒")
```

```python
# 區間
between_percent = [0.4, 0.6]

y_between = y[int(len(y)*between_percent[0]):int(len(y)*between_percent[1])]
speech_onnx_true_between = speech_onnx_true[int(len(speech_onnx_true)*between_percent[0]):int(len(speech_onnx_true)*between_percent[1])]
speech_onnx_false_between = speech_onnx_true[int(len(speech_onnx_false)*between_percent[0]):int(len(speech_onnx_false)*between_percent[1])]

# 创建一个包含三个子图的图表
plt.figure(figsize=(12, 12))

# 第一个子图：音频振幅
plt.subplot(3, 1, 1)
plt.plot(y_between)
plt.title('Audio Amplitude')
plt.xlabel('Sample')
plt.ylabel('Amplitude')

# 第二个子图：VAD概率 ONNX True
plt.subplot(3, 1, 2)
plt.fill_between(np.arange(len(speech_onnx_true_between)), speech_onnx_true_between, alpha=0.4)
plt.plot(speech_onnx_true_between, alpha=0.6)
plt.title('VAD Probabilities ONNX True')
plt.xlabel('Sample')
plt.ylabel('Probability')

# 第三个子图：VAD概率 ONNX False
plt.subplot(3, 1, 3)
plt.fill_between(np.arange(len(speech_onnx_false_between)), speech_onnx_false_between, alpha=0.4)
plt.plot(speech_onnx_false_between, alpha=0.6)
plt.title('VAD Probabilities ONNX False')
plt.xlabel('Sample')
plt.ylabel('Probability')

# 显示图表
plt.tight_layout()
plt.show()
```
